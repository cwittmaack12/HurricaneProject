# Univariate Analysis

```python
plt.figure(figsize=(12,6))

sns.violinplot(x=df['avg_temp'])

plt.xlabel("Temp", fontsize=14)
plt.title("Average Temp Distribution");
```

```python
sns.displot(data = df, x = 'temp_v_avg', kind = 'hist', bins=20, kde=True)

plt.xlabel("Degrees", fontsize=14)
plt.ylabel("Number of Years", fontsize=14)
plt.title("Distribution of Variance From Average Temp", fontsize=14);
```

### This is likely misleading.  Need to look at it in time series because the bulk of the negative value figures were decades ago whereas most of the positive values were in the last few decades.

```python
plt.figure(figsize=(12,6))

ax = sns.histplot(data=df, x='global_landfall_hurricane', palette="viridis", bins=10)

# sets x and y labels
plt.xlabel("Number of Landfall Hurricanes", fontsize=14)
plt.ylabel("Occurrences", fontsize=14)

# sets title
plt.title("Distribution of Landfall Hurricanes per Year", fontsize=15);
```

```python
sns.displot(data = df, x = 'global_landfall_strong', kind = 'hist', bins=10)

plt.xlabel("Strong Storms", fontsize=14)
plt.ylabel("Occurrences", fontsize=14)
plt.title("Distribution of Strong Landfall Storms", fontsize=14);
```

```python
sns.displot(data = df, x = 'global_landfall_weak', kind = 'hist', bins=15)

plt.xlabel("Weak Storms", fontsize=14)
plt.ylabel("Occurrences", fontsize=14)
plt.title("Distribution of Weak Landfall Storms", fontsize=14);
```

```python
plt.figure(figsize=(12,6))

ax = sns.countplot(data=df, x='total_hurricanes',palette="icefire")

plt.xticks(rotation=90)
plt.xlabel("Number of Hurricanes per Year", fontsize=13)
plt.ylabel("Number of Occurrences", fontsize=13)
plt.title("Prevalence of Hurricanes per Year (1895-2021)", fontsize=15);
```

```python
plt.figure(figsize=(12,6))

ax = sns.countplot(data=df, x='total_major_hurricanes',palette="icefire")

plt.xticks(rotation=90)
plt.xlabel("Number of Major Hurricanes per Year", fontsize=13)
plt.ylabel("Number of Occurrences", fontsize=13)
plt.title("Prevalence of Major Hurricanes per Year (1895-2021)", fontsize=15);
```

```python
plt.figure(figsize=(12,6))

ax = sns.histplot(data=df, x='tornadoes', bins=30)

plt.xlabel("Number of Tornadoes", fontsize=13)
plt.ylabel("Occurrences", fontsize=13)
plt.title("Distribution of Tornadoes per Year (1950-2021)", fontsize=16);
```

```python
plt.figure(figsize=(12,6))

sns.boxplot(x=df['tornadoes'])

plt.xlabel("Number of Tornadoes", fontsize=14)
plt.title("Distribution of Tornadoes per Year (1950-2021)");
```

```python
plt.figure(figsize=(12,6))

sns.boxplot(x=df['num_wildfires'])

plt.xlabel("Number of Wildfires", fontsize=14)
plt.title("Distribution of Wildfires per Year (1983-2021)");
```

```python
plt.figure(figsize=(12,6))

sns.boxplot(x=df['acres_burned']/1000000)


plt.title("Distribution of Acres Burned")
plt.xlabel("Acres Burned in Millions (1983-2021)", fontsize=14)
plt.ticklabel_format(style='plain', axis='x')
```

# Bivariate Relationships

### Average Temp

```python
temp = df.avg_temp
average = temp.rolling(window=7).mean()
x=df.year
```

```python
plt.figure(figsize=(18, 8))
plt.plot(x,temp, marker = 'H', linewidth=2)
plt.plot(x,average)
plt.ylabel('Temperature (deg F)')
```

```python
plt.figure(figsize=(12,4))
sns.lineplot(data=df, x='year', y='avg_temp', marker='h', markersize=6 )
```

```python
sns.lmplot(x='year', y='avg_temp', data=df, aspect=3)
```

### Temp v Average

```python
plt.figure(figsize=(12,4))
sns.lineplot(data=df, x='year', y='temp_v_avg', marker='h', markersize=6 )
```

```python
sns.lmplot(x='year', y='temp_v_avg', data=df, aspect=3)
```

```python
temp = df.temp_v_avg
average = temp.rolling(window=7).mean()
x=df.year
```

```python
plt.figure(figsize=(18, 8))
plt.plot(x,temp, marker = 'H', linewidth=2)
plt.plot(x,average)
plt.ylabel('Temperature (deg F)')
```

### Global Landfall Hurricanes

```python
lf=df.iloc[55:,:]
plt.figure(figsize=(12, 8))
sns.lmplot(x='year', y='global_landfall_hurricane', data=lf, aspect=3)
```

```python
plt.figure(figsize=(18, 8))
stack_bar=df.iloc[55:,:]
sns.barplot(data=stack_bar,x='year', y='global_landfall_hurricane', color='gray')
sns.barplot(data=stack_bar,x='year', y='global_landfall_strong', color='yellow')
plt.legend(labels=['legendEntry1', 'legendEntry2'])
```
